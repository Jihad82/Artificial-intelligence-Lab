```python
class N_queen:
    def __init__(self, a):
        self.N = a
        self.solutions = []
    
    def printSolution(self, board):
        for i in range(self.N):
            for j in range(self.N):
                print(" " + str(board[i][j]) + " ", end="")
            print()
    
    def isSafe(self, grid, row, col):
        for i in range(col):
            if grid[row][i] == 1:
                return False
        i = row
        j = col
        while i >= 0 and j >= 0:
            if grid[i][j] == 1:
                return False
            i -= 1
            j -= 1
        i = row
        j = col
        while j >= 0 and i < self.N:
            if grid[i][j] == 1:
                return False
            i += 1
            j -= 1
        return True
    
    def solveNQUtil(self, grid, col):
        if col >= self.N:
            temp = []
            for i in range(self.N):
                temp.append(grid[i][:])
            self.solutions.append(temp)
            return True
        for i in range(self.N):
            if self.isSafe(grid, i, col):
                grid[i][col] = 1
                self.solveNQUtil(grid, col + 1)
                grid[i][col] = 0
    
    def solveNQ(self):
        grid = [[0 for x in range(self.N)] for y in range(self.N)]
        self.solveNQUtil(grid, 0)
        num_solutions = len(self.solutions)
        if num_solutions == 0:
            print("No solution exists for " + str(self.N) + " queens")
            return False
        print("Number of solutions found for " + str(self.N) + " queens: " + str(num_solutions))
        for solution in self.solutions:
            self.printSolution(solution)
            print()
        return True

n = int(input("Number of queens to place: "))
Queen = N_queen(n)
Queen.solveNQ()
```
